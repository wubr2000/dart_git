###Elbow plot for determining optimal number of clusters

```python
#Assign a range of clusters for testing
k_range = range(1,37,4) #range(start,end,increments)
```

```python
#Fit the Mini-batch K-Means model for each n_cluster = k
k_means_var = [MiniBatchKMeans(init='k-means++', n_clusters=k).fit(df_final) for k in k_range]
```

```python
#Pull out the cluster centers for each model
centroids = [X.cluster_centers_ for X in k_means_var]
```

```python
#Calculate "within-cluster" sum of squares
D_k = [cdist(df_final, cent, 'euclidean') for cent in centroids]
```

```python
cIdx = [np.argmin(D,axis=1) for D in D_k]
```

```python
dist = [np.min(D,axis=1) for D in D_k]
```

```python
avgWithinSS = [sum(d**2)/df_final.shape[0] for d in dist]
```

```python
#Elbow Curve
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(k_range[1:], avgWithinSS[1:], 'b*-')
ax.plot(k_range[5], avgWithinSS[5], marker='o', markersize=12, 
    markeredgewidth=2, markeredgecolor='r', markerfacecolor='None')
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('Average within-cluster sum of squares')
plt.title('Elbow for KMeans clustering')
```

###Silhouette coefficient plot for determining optimal number of clusters

1. Silhouette coefficient (this takes a long time for >50 clusters)
1. No "ground truth" needed, i.e. no labeling needed
1. Basically calculates how "dense" the clusters are - ideally, clusters should be quite dense and not overlap.
1. The score is bounded between -1 for incorrect clustering and +1 for highly dense clustering. 
1. Scores around zero indicate overlapping clusters.

```python
#Sample when data size becomes too large (n>10000 or so) - pairwise distance calc so can't take too many data points
#sample_rows = np.random.choice(df_final.index, 5000)
#df_test = df_final.ix[sample_rows]

df_test = df_final
df_test.shape
```

```python
#Assign a range of clusters for testing
k_range = range(5,26,5) #range(start,end,increments)
```

```python
#Fit the Mini-batch K-Means model for each n_cluster = k
k_means_var = [MiniBatchKMeans(init='k-means++', n_clusters=k).fit(df_test) for k in k_range]
```

```python
#Pull out the cluster labels for each model
k_means_labels = [X.labels_ for X in k_means_var]
```

```python
#Calculate Silhouette Coefficient for each n_cluster in k_range (THIS TAKES A WHILE!)
silhouette = [metrics.silhouette_score(df_test, labels, metric='euclidean') for labels in k_means_labels]
```

```python
#Silhouette Coefficient curve
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(k_range[:], silhouette[:], 'b*-')
ax.plot(k_range[3], silhouette[3], marker='o', markersize=12, 
    markeredgewidth=2, markeredgecolor='r', markerfacecolor='None')
ax.set_ylim((0.5,0.9))
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Coefficient (between -1 and 1)')
plt.title('Silhouette Coefficients for KMeans clustering')
```

###Plotting K-means cluster onto 2-D PCA plot

```python
###############################################################################
# Visualize the results on PCA-reduced data

#Sample when dataset is too large (n>50,000)
#sample_rows = np.random.choice(df_final.index, 30000)
#X = df_final.ix[sample_rows].values.astype(float)

X = df_final.values
data = scale(X)
reduced_data = PCA(n_components=2).fit_transform(data)
kmeans = KMeans(init='k-means++', n_clusters=20)
km = kmeans.fit(reduced_data)
```

```python
# Step size of the mesh. Decrease to increase the quality of the VQ.
h = 0.2     # point in the mesh [x_min, m_max]x[y_min, y_max].

# Plot the decision boundary. For that, we will assign a color to each
x_min, x_max = reduced_data[:, 0].min() + 1, reduced_data[:, 0].max() - 1
y_min, y_max = reduced_data[:, 1].min() + 1, reduced_data[:, 1].max() - 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point in mesh. Use last trained model.
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Paired,
           aspect='auto', origin='lower')

plt.plot(reduced_data[:, 0], reduced_data[:, 1], 'k.', markersize=2)
# Plot the centroids as a white X
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='x', s=169, linewidths=3,
            color='w', zorder=10)
plt.title('K-means clustering on the digits dataset (PCA-reduced data)\n'
          'Centroids are marked with white cross')
plt.xlim(x_min+200, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```

###Parsing the Customer Target Query tag and Featurizing

```python
raw_tags = dart_has_audience_id1['custom_targeting_query']
```

```python
#Clean the raw tags before breaking them out into key/value pairs
cleaned_tags = [raw_tag.replace("{","").replace("}","").replace('"',"").split("],") for raw_tag in raw_tags]
```

```python
#Put cleaned tags into a key/value pair list for each user
customer_targeting_query = []
for user, kv in enumerate(cleaned_tags):
    key_value = []
    for pair in kv:
        key_value.append(pair.replace("]","").split(":["))
    customer_targeting_query.append(key_value)
```

```python
#Extract all unique keys from key/value pair for featurization
unique_keys = list(set([pair[0] for user in customer_targeting_query for pair in user]))
print "Number of unique keys from key/value pair:", len(unique_keys)
```

```python
#Create a dummy variable column for every KEY in unique customer targeting query tag. 
#Initialize each cell to a blank (for strings input).
for unique_key in unique_keys:
    dart_has_audience_id1[unique_key] = ""

dart_has_audience_id1.shape
```

```python
#Create Dataframe with Customer Query Keys Columns Only
keys_beg_col = dart_has_audience_id1.shape[1] - len(unique_keys)
dart_customer_query_keys_only =  dart_has_audience_id1.iloc[:,keys_beg_col:]
```

```python
#For each user, put in applicable values in each key column:
for user, kv_pair in enumerate(customer_targeting_query):
    #Counter to check how many users it has looped through
    if (user+1)%5000 == 0:
        print user+1
    #Put in values for each applicable key column
    for k, v in kv_pair:
        dart_customer_query_keys_only.iloc[user][k] = str(v)
```

```python
dart_customer_query_keys_only.head(2)
```

```python
#Use get_dummies() to featurize customer query keys
df_customer_queries = []
key_value_pairs = []
for unique_key in unique_keys:
    print unique_key,
    df_intermediate = dart_customer_query_keys_only[unique_key].str.get_dummies(sep=",")
    #Get every key/value pair for column header later
    for col in df_intermediate.columns:
        key_value_pairs.append(unique_key+'-'+col)
    #Append key to each key-value column
    df_customer_queries.append(df_intermediate)
```

```python
#Concatenating all customer targeting query tag columns
df_customer_query_featurized = pd.concat([df_customer_queries[num] for num in xrange(0,len(unique_keys)-1)], axis= 1)
#Replace header with name of every key/value pair
df_customer_query_featurized.columns = key_value_pairs

df_customer_query_featurized.shape
```
