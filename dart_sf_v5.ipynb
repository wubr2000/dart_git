###Featurizing Customer Target Query Tags and Audience Segment IDs

```python
dart_has_audience_id.shape
```

```python
#Sample rows for faster analysis
#RESET_INDEX AND DROP AND COPY IS IMPORTANT - otherwise it retains the entire dataframe and so will still be slow
num_of_rows_to_sample = 30000

sample_rows = np.random.choice(dart_has_audience_id.index, num_of_rows_to_sample)
dart_has_audience_id1 = dart_has_audience_id.ix[sample_rows].reset_index(drop=True).copy()
```

###Parsing the Customer Target Query tag and Featurizing

```python
#Clean the raw customer targeting query tags 
#and split them out into key/value pairs
raw_tags = dart_has_audience_id1['custom_targeting_query']
cleaned_tags = [raw_tag.replace("{","").replace("}","").replace('"',"").split("],") for raw_tag in raw_tags]
```

```python
#Put cleaned tags into a key/value pair list for each user
customer_targeting_query = []
for user, kv in enumerate(cleaned_tags):
    key_value = []
    for pair in kv:
        key_value.append(pair.replace("]","").split(":["))
    customer_targeting_query.append(key_value)
```

```python
#Extract all unique keys from key/value pair for featurization
unique_keys = list(set([pair[0] for user in customer_targeting_query for pair in user]))
print "Number of unique keys from key/value pair:", len(unique_keys)
```

```python
#Create a dummy variable column for every KEY in unique customer targeting query tag. 
#Initialize each cell to a blank (for strings input).
dart_has_audience_id2 = dart_has_audience_id1.copy()

for unique_key in unique_keys:
    dart_has_audience_id2[unique_key] = ""

dart_has_audience_id2.shape
```

```python
#Create Dataframe with Customer Query Keys Columns Only
keys_beg_col = dart_has_audience_id2.shape[1] - len(unique_keys)
dart_customer_query_keys_only =  dart_has_audience_id2.iloc[:,keys_beg_col:]
```

```python
#For each user, put in applicable values in each key column:
for user, kv_pair in enumerate(customer_targeting_query):
    #Counter to check how many users it has looped through
    if (user+1)%5000 == 0:
        print user+1,
    #Put in values for each applicable key column
    for k, v in kv_pair:
        dart_customer_query_keys_only.iloc[user][k] = str(v)
```

```python
dart_customer_query_keys_only.head(2)
```

```python
#Use get_dummies() to featurize all customer query key/value pairs
df_customer_queries = []
key_value_pairs = []
for unique_key in unique_keys:
    print unique_key,
    df_intermediate = dart_customer_query_keys_only[unique_key].str.get_dummies(sep=",")
    #Get every key/value pair for column header later
    for col in df_intermediate.columns:
        key_value_pairs.append(unique_key+'-'+col)
    #Append key to each key-value column
    df_customer_queries.append(df_intermediate)
```

```python
#Concatenate all featurized dataframes 
df_customer_query_featurized = pd.concat([df_customer_queries[num] for num in xrange(0,len(unique_keys))], axis= 1)
#Replace header with names of every key/value pair
df_customer_query_featurized.columns = key_value_pairs

df_customer_query_featurized.shape
```

```python
df_customer_query_featurized.head(2)
```

###Parsing the Audience Segment IDs and Featurizing

```python
#Spliting the audience segment ID for each user and storing this into a list
aud_id = [id.split("|") for id in dart_has_audience_id1["audiencesegmentids"]]
aud_seg = [i for lst in aud_id for i in lst]
aud_features = list(set(aud_seg))
```

```python
#Create dummy variable columns for each audience segment ID. Initialize each cell to 0.
for feature in aud_features:
    dart_has_audience_id1[feature] = 0
    
dart_has_audience_id1.shape
```

```python
dart_has_audience_id1.head(2)
```

```python
#For each user, put a "1" in the applicable audience segment ID column
for k, v in enumerate(aud_id):
    for singleID in v:
        dart_has_audience_id1[singleID][k] = 1
```

```python
#Drop the audiencesegmentids column now that it has been encoded as individual feature columns
del dart_has_audience_id1["audiencesegmentids"]
```

```python
dart_has_audience_id1.shape
```

```python
dart_has_audience_id1.head(2)
```

###Final clean-up of dataframe before analysis

```python
#Initialize dataframe for KMeans analysis
dart_intermediate1 = dart_has_audience_id1.copy()

#Drop columns for analysis
df_final1 = dart_intermediate1.drop(["time","time_gmt","creativesize","logdate_gmt","creativeversion","custom_targeting_query","domain","postalcodeid",
                  "browser","os","osversion","timeusec","requestedadunitsizes","mobiledevice","mobilecapability",
                  "mobilecarrier","iscompanion","publisherprovidedid","logdate","hour","source_file",
                  "country","region","metro","city","bandwidth","product"], axis=1)

#Replace all non-digits and "NA"s in these columns with 0
cols_to_replace = ['bandwidthgroupid','postalcode','osid','countryid','regionid','metroid','cityid','browserid']
for col in cols_to_replace:
    df_final1[col].fillna(0, inplace=True)
    df_final1[col].replace(to_replace='[^0-9]+', value=0,inplace=True,regex=True) 
```

```python
#Check for any NaN values
df_final1.isnull().sum().sum()
```

```python
#Drop columns that are the same for every user in this regional subset:
df_final1 = df_final1.drop(['countryid','regionid','metroid'],axis=1)
```

```python
df_final1.shape
```

```python
df_segments_tags = df_final1.iloc[:,11:]
df_segments_tags["userid"] = df_final1["userid"]

df_segments_tags.head(2)
```

###Combining Customer Targeting Queries and Audience Segment ID Dataframes

```python
#Combine two large sparse dataframes
#This takes a while
df_queries_and_segments = pd.concat([df_customer_query_featurized, df_segments_tags], axis=1)
```

##Random Forest - First Run

```python
print df_segments_tags.shape
print df_customer_query_featurized.shape
print df_queries_and_segments.shape
```

```python
#Unique users and prices for homes in the listing detail pages (LDP) they've viewed
all_pldp = [headers for headers in df_queries_and_segments.columns.values.tolist() if bool(re.match(r'pldp*', headers))]
pd.DataFrame(df_queries_and_segments.groupby(["userid"])[all_pldp].sum().sum(), columns=["count"]).sort(["count"]).plot(kind="bar",
                                                                                                                        title = "Histogram for price ranges of homes on LDP for Unique Users")
```

```python
#Setting hyper-parameters for random forest classifier
RF = RandomForestClassifier(
                            criterion = "entropy",
                            n_estimators = 500,
                            max_features = 'sqrt',
                            oob_score = True,
                            max_depth = 3,
                            #n_jobs = -1,
                            verbose = 1
)
```

###RF for users who have viewed a home (in LDP) worth > 1mm

```python
#Price of homes viewed on Landing Detail Page > 1mm
above_1mm = ['pldp-1mm','pldp-1mm_5mm','pldp-5mm']
```

```python
#Create a Y-labels for all LDP prices > 1mm
Y = df_customer_query_featurized[above_1mm].sum(axis=1)

#Only use audience segment IDs for features:
X = df_segments_tags[aud_features]

print Y.shape
print X.shape
```

```python
start = time()
rf_dart = RF.fit(X,Y)
print "training time:", time() - start
```

```python
#Feature importance
important_features_id = df_segments_tags.columns[np.argsort(rf_dart.feature_importances_)[::-1]]
```

```python
#Read in audience segment IDs description table
dart_aud = pd.read_csv('dart_aud.csv', sep=',')
```

```python
print "Top 20 'Most Significant' DART Audience Segments For Predicting Who Viewed Homes on Landing Page > 1mm"
print "======================================================================================================"
for important in important_features_id[:20]:
    print dart_aud[dart_aud['id']==important]["name"].values
```

```python
rf_dart.oob_score_
```
