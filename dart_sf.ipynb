#Encoding Categorical Variables (in particular, Audience Segment IDs in DART data)

```python
dartdf = pd.read_csv('dart_sf.csv', sep=',')
```

```python
dartdf.head()
```

```python
#Only use rows that have audience segments ids
dart_has_audience_id = dartdf[dartdf["audiencesegmentids"] != "\N"].reset_index(drop=True)
```

```python
dart_has_audience_id.shape
```

```python
#Spliting the audience segment ID for each user and storing this into a list
aud_id = [id.split("|") for id in dart_has_audience_id["audiencesegmentids"]]
aud_seg = [i for lst in aud_id for i in lst]
aud_features = list(set(aud_seg))

len(aud_id)
```

#K-means Analysis

```python
#Initialize dataframe for KMeans analysis
dart_kmeans = dart_has_audience_id.copy()
```

```python
#Drop columns for analysis
df_final = dart_kmeans.drop(["time","time_gmt","userid","creativesize","logdate_gmt","creativeversion","custom_targeting_query","domain","postalcodeid",
                  "browser","os","osversion","timeusec","requestedadunitsizes","mobiledevice","mobilecapability",
                  "mobilecarrier","iscompanion","publisherprovidedid","logdate","hour","source_file","Unnamed: 42",
                  "country","region","metro","city","bandwidth","product"], axis=1)
```

```python
#Replace all non-digits and "\N" in the following columns with 0

cols_to_replace = ['bandwidthgroupid','postalcode','osid','countryid','regionid','metroid','cityid','browserid']
for col in cols_to_replace:
    df_final[col].replace(to_replace='[^0-9]+', value=0,inplace=True,regex=True) 
```

```python
df_final.head()
```

```python
#K-Means Model Parameters
#Use MiniBatchKmeans - ALOT faster for n>10,000 samples
#For reasons: http://algorithmicthoughts.wordpress.com/2013/07/26/machine-learning-mini-batch-k-means/

n_clusters = 20
batch_size = 100 #Default
n_init = 300 #default=3. Number of random initializations that are tried. 
           #In contrast to KMeans, the algorithm is only run once, 
           #using the best of the n_init initializations as measured by inertia.

mbkmeans = MiniBatchKMeans(init='k-means++', 
                         n_clusters=n_clusters, 
#                          batch_size=batch_size,
#                          n_init=n_init,
#                          max_no_improvement=10, 
                         verbose=0)
```

```python
#Fitting the K-means model to data
mbkmeans_model = mbkmeans.fit(df_final)
labels = mbkmeans_model.labels_
```

###Elbow plot for determining optimal number of clusters

```python
#Assign a range of clusters for testing
k_range = range(1,37,4) #range(start,end,increments)
```

```python
#Fit the Mini-batch K-Means model for each n_cluster = k
k_means_var = [MiniBatchKMeans(init='k-means++', n_clusters=k).fit(df_final) for k in k_range]
```

```python
#Pull out the cluster centers for each model
centroids = [X.cluster_centers_ for X in k_means_var]
```

```python
#Calculate "within-cluster" sum of squares
D_k = [cdist(df_final, cent, 'euclidean') for cent in centroids]
```

```python
cIdx = [np.argmin(D,axis=1) for D in D_k]
```

```python
dist = [np.min(D,axis=1) for D in D_k]
```

```python
avgWithinSS = [sum(d**2)/df_final.shape[0] for d in dist]
```

```python
#Elbow Curve
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(k_range[1:], avgWithinSS[1:], 'b*-')
ax.plot(k_range[5], avgWithinSS[5], marker='o', markersize=12, 
    markeredgewidth=2, markeredgecolor='r', markerfacecolor='None')
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('Average within-cluster sum of squares')
plt.title('Elbow for KMeans clustering')
```

###Silhouette coefficient plot for determining optimal number of clusters

1. Silhouette coefficient (this takes a long time for >50 clusters)
1. No "ground truth" needed, i.e. no labeling needed
1. Basically calculates how "dense" the clusters are - ideally, clusters should be quite dense and not overlap.
1. The score is bounded between -1 for incorrect clustering and +1 for highly dense clustering. 
1. Scores around zero indicate overlapping clusters.

```python
sample_rows = np.random.choice(df_final.index, 5000)
df_test = df_final.ix[sample_rows]
df_test.shape
```

```python
#Assign a range of clusters for testing
k_range = range(5,26,5) #range(start,end,increments)
```

```python
#Fit the Mini-batch K-Means model for each n_cluster = k
k_means_var = [MiniBatchKMeans(init='k-means++', n_clusters=k).fit(df_test) for k in k_range]
```

```python
#Pull out the cluster labels for each model
k_means_labels = [X.labels_ for X in k_means_var]
```

```python
#Calculate Silhouette Coefficient for each n_cluster in k_range (THIS TAKES A WHILE!)
silhouette = [metrics.silhouette_score(df_test, labels, metric='euclidean') for labels in k_means_labels]
```

```python
#Silhouette Coefficient curve
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(k_range[:], silhouette[:], 'b*-')
ax.plot(k_range[3], silhouette[3], marker='o', markersize=12, 
    markeredgewidth=2, markeredgecolor='r', markerfacecolor='None')
ax.set_ylim((0.5,1))
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Coefficient (between -1 and 1)')
plt.title('Silhouette Coefficients for KMeans clustering')
```

###Plotting K-means cluster onto 2-D PCA plot

```python
###############################################################################
# Visualize the results on PCA-reduced data

sample_rows = np.random.choice(df_final.index, 30000)
X = df_final.ix[sample_rows].values.astype(float)
data = scale(X) 
reduced_data = PCA(n_components=2).fit_transform(data)
mbkmeans = MiniBatchKMeans(init='k-means++', n_clusters=20)
mbkm = mbkmeans.fit(reduced_data)
```

```python
# Step size of the mesh. Decrease to increase the quality of the VQ.
h = .02     # point in the mesh [x_min, m_max]x[y_min, y_max].

# Plot the decision boundary. For that, we will assign a color to each
x_min, x_max = reduced_data[:, 0].min() + 1, reduced_data[:, 0].max() - 1
y_min, y_max = reduced_data[:, 1].min() + 1, reduced_data[:, 1].max() - 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point in mesh. Use last trained model.
Z = mbkmeans.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1)
plt.clf()
plt.imshow(Z, interpolation='nearest',
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Paired,
           aspect='auto', origin='lower')

plt.plot(reduced_data[:, 0], reduced_data[:, 1], 'k.', markersize=2)
# Plot the centroids as a white X
centroids = mbkmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='x', s=169, linewidths=3,
            color='w', zorder=10)
plt.title('K-means clustering on the digits dataset (PCA-reduced data)\n'
          'Centroids are marked with white cross')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```
